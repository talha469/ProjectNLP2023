```python
import nltk
import re
nltk.download('gutenberg')

from nltk.corpus import gutenberg

# Download the "blake-poems" and "austen-emma" texts
blake_poems = gutenberg.raw('blake-poems.txt')
austen_emma = gutenberg.raw('austen-emma.txt')

def separate_poem_into_chapters(poem):
    chapters = re.split(r'\n \n(?=[A-Z ]+\n)', blake_poems)
    return chapters[2:]

blake_chapters = separate_poem_into_chapters(blake_poems)

austen_chapters = separate_poem_into_chapters(austen_emma)
```

```python
from collections import Counter
from scipy.stats import kurtosis
import matplotlib.pyplot as plt
import numpy as np

# Tokenize the text into lines
def tokenize_lines(text):
    return nltk.sent_tokenize(text)

# Analyze repetitions within lines and between successive lines
def analyze_repetitions(lines):
    line_repetitions = []
    successive_repetitions = []

    for i in range(1, len(lines)):
        line1 = lines[i - 1]
        line2 = lines[i]

        # Tokenize lines into words
        words1 = re.findall(r'\b\w+\b', line1.lower())
        words2 = re.findall(r'\b\w+\b', line2.lower())

        count1 = Counter(words1)
        count2 = Counter(words2)

        repetition_count = sum((count1 & count2).values())
        line_repetitions.append(repetition_count)

        successive_repetition_count = sum((count1 & count2).values())
        successive_repetitions.append(successive_repetition_count)

    return line_repetitions, successive_repetitions

# Calculate the mean, standard deviation, and kurtosis of repetitions per chapter
def calculate_statistics(repetitions):
    mean = np.mean(repetitions)
    std_dev = np.std(repetitions)
    kurt = kurtosis(repetitions)
    return mean, std_dev, kurt

# Create a distribution plot
def plot_distribution(repetitions, title):
    plt.figure(figsize=(10, 6))
    plt.hist(repetitions, bins=20, color='blue', edgecolor='black')
    plt.title(title)
    plt.xlabel('Number of Repetitions')
    plt.ylabel('Frequency')
    plt.show()

for i, chapter in enumerate(blake_chapters):
    lines = tokenize_lines(chapter)
    line_repetitions, successive_repetitions = analyze_repetitions(lines)
    
    mean, std_dev, kurt = calculate_statistics(line_repetitions)
    print(f"Blake's Chapter {i + 1} Line Repetitions:")
    print(f"Mean: {mean}, Standard Deviation: {std_dev}, Kurtosis: {kurt}")
    plot_distribution(line_repetitions, f"Blake's Chapter {i + 1} Line Repetitions")

for i, chapter in enumerate(austen_chapters):
    lines = tokenize_lines(chapter)
    line_repetitions, successive_repetitions = analyze_repetitions(lines)
    
    mean, std_dev, kurt = calculate_statistics(line_repetitions)
    print(f"austen's Chapter {i + 1} Line Repetitions:")
    print(f"Mean: {mean}, Standard Deviation: {std_dev}, Kurtosis: {kurt}")
    plot_distribution(line_repetitions, f"austen's Chapter {i + 1} Line Repetitions")
```
